```python
import numpy as np
import matplotlib.pyplot as plt

data_laplace = np.loadtxt('../build/laplace_result.txt')
data_helmholtz = np.loadtxt('../build/helmholtz_result.txt')
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.title('real part of solution')
plt.plot(data_laplace[:,0])
plt.plot(data_helmholtz[:,0])
plt.legend(['laplace', 'helmholtz'])
plt.subplot(1, 2, 2)
plt.title('imaginary part of solution')
plt.plot(data_laplace[:,1])
plt.plot(data_helmholtz[:,1])
plt.legend(['laplace', 'helmholtz'])
plt.show()
```
